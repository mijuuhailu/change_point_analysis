# Task 1: Laying the Foundation for Analysis

## Project Overview

### Objective

The objective of this project is to analyze historical Brent crude oil prices and identify significant structural changes associated with major geopolitical events, OPEC policy decisions, economic crises, and global conflicts. Bayesian Change Point Detection will be used to detect changes in the statistical behavior of oil prices, and the detected change points will be compared with historical events to identify possible explanations for these shifts.

---

# 1. Data Analysis Workflow

The analysis will follow the workflow below:

1. Load and inspect the Brent oil price dataset.
2. Clean and preprocess the data.
3. Perform exploratory data analysis (EDA).
4. Analyze key time series properties (trend, stationarity, and volatility).
5. Compile a dataset of major geopolitical and economic events.
6. Build a Bayesian Change Point Detection model using PyMC.
7. Identify significant structural breaks in the price series.
8. Compare detected change points with historical events.
9. Interpret the statistical and economic significance of the results.
10. Present findings through visualizations and an interactive dashboard.

---

# 2. Event Dataset

A structured dataset (`events.csv`) was created containing major events that may have influenced Brent oil prices.

The dataset includes:

- Date
- Event description
- Event category (Conflict, OPEC, Economy, Sanctions, etc.)

These events will later be compared with the detected change points.

---

# 3. Assumptions

The analysis is based on the following assumptions:

- Brent crude oil prices reflect major global oil market conditions.
- Historical event dates are approximately correct.
- Market reactions occur close to the reported event dates.
- Structural changes detected by the model may be associated with major events.

---

# 4. Limitations

This analysis has several limitations:

- Correlation does not imply causation.
- Multiple events may occur around the same time.
- Financial markets often react before official announcements.
- Other important economic variables (GDP, inflation, exchange rates, demand, etc.) are not included.
- The Bayesian model detects statistical changes rather than proving that a specific event caused the change.

# 6. Understanding the Data

Before modeling, the time series will be analyzed to understand its statistical properties.

## Trend

Trend analysis identifies the long-term movement of Brent oil prices over time. Detecting an increasing or decreasing trend helps determine whether the series should be transformed before modeling.

## Stationarity

Stationarity means that the statistical properties of a time series remain constant over time. Since raw oil prices are generally non-stationary, log returns will also be analyzed because they are typically closer to stationarity and are more suitable for statistical modeling.

## Volatility

Oil prices often experience periods of calm followed by periods of large fluctuations. This phenomenon, known as volatility clustering, is common in financial time series and should be considered when interpreting model results.

---

# 7. Change Point Models

A change point model identifies points in time where the statistical behavior of a time series changes significantly.

In this project, Bayesian Change Point Detection will estimate:

- the most probable change point,
- the average oil price before the change,
- the average oil price after the change,
- and the uncertainty associated with these estimates.

These detected changes will then be compared with historical geopolitical and economic events.
